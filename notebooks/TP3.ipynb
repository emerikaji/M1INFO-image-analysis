## Differentiator: Sharpening a image

We propose to filter the _cameraman.png_ image by using this second kernel: 

\begin{equation*}
h_2(i,j) = 
\begin{pmatrix}
0 & -1 & 0 \\
-1 & \bf{5} & -1 \\
0 & -1 & 0  
\end{pmatrix}
\end{equation*}

In your python code,  load of the image, filter the image, and display the images (original and filtered)

**Questions:**

- why the output image is sharpened?
- what happens if the gain of the kernel equals 0? (define the new kernel and experiment it)

**Responses:**


```python
 
# RAZ (figures closing)
plt.close('all')

# image reading
im = plt.imread('../standard_test_images/cameraman.png')


# ... and normalization (if necessary)
im = im / im.max()
im = 255 * im
im = im.astype(np.uint8)

# apply kernel
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
im2 = cv.filter2D(im, -1, kernel) 

# display    
plt.figure()
plt.imshow(im, cmap="gray")
plt.colorbar()
plt.show()

plt.figure()
plt.imshow(im2, cmap="gray")
plt.colorbar()
plt.show()
```
